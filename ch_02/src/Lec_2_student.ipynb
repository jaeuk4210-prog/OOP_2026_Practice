### 1-1. Dynamic Typing (Slide 10)

Python is a **dynamically typed** language.  
You can reassign values of different types to the same variable.

````python
# Dynamic typing demo
x = 42
print(f"x = {x}, type = {type(x)}")  # int

x = "Hello"
print(f"x = {x}, type = {type(x)}")  # str

x = [1, 2, 3]
print(f"x = {x}, type = {type(x)}")  # list
````

### 1-2. `type()` and `isinstance()` (Slide 10)

- `type(obj)` — Returns the exact type of the object
- `isinstance(obj, cls)` — Checks the type considering inheritance

````python
# type() vs isinstance()
value = 3.14
print(type(value))                    # <class 'float'>
print(isinstance(value, float))       # True
print(isinstance(value, (int, float)))  # True — can check multiple types

# bool is a subclass of int
print(type(True))                     # <class 'bool'>
print(isinstance(True, int))          # True — reflects inheritance
````

### 1-3. Control Flow: if / elif / else (Slide 11)

Example of the `check_status` function from the slides.

````python
def check_status(score: int) -> str:
    """Returns a grade based on the score."""
    if score >= 90:
        return "A — Excellent"
    elif score >= 80:
        return "B — Good"
    elif score >= 70:
        return "C — Average"
    else:
        return "F — Needs Improvement"


# Test
for s in [95, 82, 73, 55]:
    print(f"score={s:3d} → {check_status(s)}")
````

## Practice 2: Function Definitions (LEGB, *args, **kwargs)
📖 **Slides 13, 56-58**

### 2-1. LEGB Scope Rules (Slide 13)

Python looks up names in the order **L → E → G → B**.

- **L**ocal: Inside the function
- **E**nclosing: Outer function (in nested functions)
- **G**lobal: Module level
- **B**uilt-in: Built-in names like `print`, `len`

````python
# LEGB scope demo
g_var = "Global"


def outer():
    e_var = "Enclosing"

    def inner():
        l_var = "Local"
        print(f"inner: l_var = {l_var}")     # Local
        print(f"inner: e_var = {e_var}")     # Enclosing
        print(f"inner: g_var = {g_var}")     # Global
        print(f"inner: len   = {len}")       # Built-in

    inner()


outer()
````

### 2-2. Default Arguments (Slide 56)

````python
def greet(name: str, greeting: str = "Hello") -> str:
    return f"{greeting}, {name}!"


print(greet("Alice"))                  # Uses default greeting
print(greet("Bob", "Good morning"))    # Custom greeting
````

### 2-3. `*args`와 `**kwargs` (Slides 57-58)

````python
# *args — Variable positional arguments
def add_all(*args):
    """Adds all given numbers together."""
    print(f"args = {args}, type = {type(args)}")
    return sum(args)


print(add_all(1, 2, 3))       # 6
print(add_all(10, 20, 30, 40))  # 100
````

````python
# **kwargs — Variable keyword arguments
def print_info(**kwargs):
    """Prints key-value pairs."""
    print(f"kwargs = {kwargs}, type = {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")


print_info(name="Alice", age=20, major="CS")
````

````python
# Using *args and **kwargs together
def flexible(a, b, *args, **kwargs):
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"kwargs={kwargs}")


flexible(1, 2, 3, 4, x=10, y=20)
````

### 2-4. Lambda 함수

````python
# lambda — Anonymous simple function
square = lambda x: x ** 2
print(square(5))  # 25

# Using lambda for sorting
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
students_sorted = sorted(students, key=lambda s: s[1], reverse=True)
print(students_sorted)  # Sorted by score in descending order
````

## Practice 3: Core Data Structures
📖 **Slides 15-20**

### 3-1. List (Slides 15-16)

- Ordered and **mutable**
- `append`, `extend`, slicing, list comprehension

````python
# List basic operations
numbers = [1, 2, 3]
numbers.append(4)          # Append to the end
print(numbers)             # [1, 2, 3, 4]

# Slicing
print(numbers[1:3])        # [2, 3]
print(numbers[::-1])       # [4, 3, 2, 1] — reversed

# List comprehension
squares = [x ** 2 for x in range(1, 6)]
print(squares)             # [1, 4, 9, 16, 25]

evens = [x for x in range(10) if x % 2 == 0]
print(evens)               # [0, 2, 4, 6, 8]
````

### 3-2. Tuple (Slide 17)

- Ordered and **immutable**
- Frequently used for unpacking

````python
# Tuple basics
point = (3, 4)
print(point[0], point[1])  # 3 4

# Immutability check
try:
    point[0] = 10
except TypeError as e:
    print(f"TypeError raised: {e}")

# Unpacking
x, y = point
print(f"x={x}, y={y}")

# Swap using tuples
a, b = 1, 2
a, b = b, a
print(f"a={a}, b={b}")  # a=2, b=1
````

### 3-3. Dictionary (Slide 18)

- **Key-value** pairs; keys must be hashable (immutable)
- `.get()`, `.items()`, `.keys()`, `.values()`

````python
# Dictionary creation and access
student = {"name": "Alice", "age": 20, "major": "CS"}
print(student["name"])          # Alice

# .get() — returns default value if key is missing (prevents KeyError)
print(student.get("gpa", "N/A"))  # N/A

# Iteration
for key, value in student.items():
    print(f"  {key}: {value}")

# Dictionary comprehension
sq_dict = {x: x ** 2 for x in range(1, 6)}
print(sq_dict)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
````

### 3-4. Set (Slide 19)

- **No duplicates**, unordered
- Set operations: intersection, union, difference

````python
# Set basics
a = {1, 2, 3, 3, 2}
print(a)  # {1, 2, 3} — duplicates removed

b = {3, 4, 5}

# Set algebra
print(f"Intersection: {a & b}")          # {3}
print(f"Union: {a | b}")                 # {1, 2, 3, 4, 5}
print(f"Difference: {a - b}")            # {1, 2}
print(f"Symmetric difference: {a ^ b}")  # {1, 2, 4, 5}

# Remove duplicates from a list
nums = [1, 1, 2, 3, 3, 3, 4]
unique = list(set(nums))
print(f"Duplicates removed: {sorted(unique)}")
````

### 3-5. Mutability vs Immutability Comparison (Slide 20)

| Type | Mutable? | Example |
|------|----------|---------|
| `int`, `float`, `str`, `tuple` | Immutable | Cannot change value |
| `list`, `dict`, `set` | Mutable | Can change value |

````python
# Mutable vs Immutable

# Immutable — string
s = "hello"
# s[0] = "H"  # TypeError!
s2 = "H" + s[1:]  # Create a new string
print(f"Original: {s}, New string: {s2}")

# Mutable — list
lst = [1, 2, 3]
lst[0] = 100  # In-place modification
print(f"Modified list: {lst}")

# Caution: aliasing with mutable objects
a = [1, 2, 3]
b = a          # Points to the same object
b.append(4)
print(f"a = {a}")  # a is also changed! [1, 2, 3, 4]
print(f"a is b: {a is b}")  # True
````

### 4-3. `__init__` Constructor and `self` (Slides 23-24)

- `__init__`: Initialization method automatically called when an instance is created
- `self`: Reference to the instance itself

````python
class Car:
    """A class representing a car."""

    def __init__(self, make: str, model: str, speed: int = 0):
        self.make = make      # Manufacturer
        self.model = model    # Model name
        self.speed = speed    # Current speed (km/h)

    def accelerate(self, amount: int) -> None:
        """Accelerates the car."""
        self.speed += amount
        print(f"{self.make} {self.model}: speed → {self.speed} km/h")

    def brake(self, amount: int) -> None:
        """Decelerates the car."""
        self.speed = max(0, self.speed - amount)
        print(f"{self.make} {self.model}: speed → {self.speed} km/h")


# Create an instance
my_car = Car("Hyundai", "Sonata")
my_car.accelerate(60)
my_car.accelerate(40)
my_car.brake(30)
print(f"Current speed: {my_car.speed} km/h")
````

### 4-4. Instance Attributes vs Class Attributes (Slide 28)

````python
class Dog:
    # Class attribute — shared by all instances
    species = "Canis familiaris"
    count = 0

    def __init__(self, name: str):
        # Instance attribute — unique to each instance
        self.name = name
        Dog.count += 1  # Modify class attribute


d1 = Dog("Buddy")
d2 = Dog("Choco")

print(f"d1.species = {d1.species}")   # Access class attribute
print(f"d1.name = {d1.name}")         # Instance attribute
print(f"d2.name = {d2.name}")
print(f"Dog.count = {Dog.count}")     # 2 — Tracking count via class attribute
````

## Practice 5: Inheritance & Polymorphism
📖 **Slides 25-29**

### 5-1. Inheritance: Animal → Dog / Cat (Slide 25)

````python
class Animal:
    """Base class for animals."""

    def __init__(self, name: str):
        self.name = name

    def speak(self) -> str:
        raise NotImplementedError("Must be implemented in a subclass")


class Dog(Animal):
    def speak(self) -> str:
        return f"{self.name}: Woof!"


class Cat(Animal):
    def speak(self) -> str:
        return f"{self.name}: Meow~"


dog = Dog("Buddy")
cat = Cat("Nabi")
print(dog.speak())
print(cat.speak())
print(f"isinstance(dog, Animal): {isinstance(dog, Animal)}")  # True
````

### 5-2. Polymorphism (Slide 26)

````python
def make_animal_speak(animal: Animal) -> None:
    """Any Animal can call speak()."""
    print(animal.speak())


# Polymorphism: same interface, different behavior
animals = [Dog("Max"), Cat("Luna"), Dog("Charlie"), Cat("Momo")]
for a in animals:
    make_animal_speak(a)
````

### 5-3. Encapsulation and Name Mangling (Slide 27)

- `_var`: Private by convention (accessible but not recommended for external use)
- `__var`: Name mangling transforms it to `_ClassName__var`

````python
class SecretAgent:
    def __init__(self, name: str, code: str):
        self.name = name
        self.__code = code  # Name mangling applied

    def reveal(self) -> str:
        return f"Agent {self.name}, code: {self.__code}"


agent = SecretAgent("Bond", "007")
print(agent.reveal())

# Direct access attempt
try:
    print(agent.__code)
except AttributeError as e:
    print(f"AttributeError: {e}")

# Accessible via name mangling (not recommended)
print(f"mangled name: {agent._SecretAgent__code}")
````

### 5-4. Introduction to Magic Methods (`__str__`, `__add__`) (Slide 29)

**Vector** class example from the slides.

````python
class Vector:
    """2D vector class — magic method demo."""

    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y

    def __str__(self) -> str:
        return f"Vector({self.x}, {self.y})"

    def __repr__(self) -> str:
        return f"Vector({self.x!r}, {self.y!r})"

    def __add__(self, other: 'Vector') -> 'Vector':
        return Vector(self.x + other.x, self.y + other.y)

    def __abs__(self) -> float:
        return (self.x ** 2 + self.y ** 2) ** 0.5


v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(v1)             # __str__ called
print(v1 + v2)        # __add__ called → Vector(4, 6)
print(abs(v1))        # __abs__ called → 5.0
print(repr(v2))       # __repr__ called
````

### 6-3. `__name__ == "__main__"` Pattern (Slide 60)

A pattern that runs specific code only when the module is executed directly.


### 6-4. Packages and `__init__.py` (Slides 63-64)

A package organizes modules into directories.

```
my_package/
├── __init__.py      # Package initializer (can be empty)
├── module_a.py
└── sub_package/
    ├── __init__.py
    └── module_b.py
```

- `__init__.py` is required for Python to recognize a directory as a package.
- Python 3.3+ supports **namespace packages** (implicit), but explicit `__init__.py` is recommended.
- Import using `from my_package import module_a` or `from my_package.sub_package import module_b`.
