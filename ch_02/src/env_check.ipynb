# Chapter 2: 환경 검증 (Environment Check)

이 노트북은 실습 환경이 올바르게 구성되었는지 확인합니다.

**확인 항목:**
1. Python 버전
2. 실행 환경 정보
3. 필수 패키지 import
4. OOP 기본 동작
5. 종합 결과


## 1. Python 버전 확인


```python
import sys

version = sys.version_info
print(f'Python 버전: {version.major}.{version.minor}.{version.micro}')
print(f'실행 경로: {sys.executable}')

assert version >= (3, 9), f'Python 3.9 이상이 필요합니다. 현재: {sys.version}'
print('✓ 버전 확인 통과 (3.9 이상)')
```

## 2. 필수 패키지 import 확인


```python
import importlib

packages = {
    'pytest'  : 'pytest',
    'bs4'     : 'beautifulsoup4',
    'PIL'     : 'Pillow',
}

for module, pkg_name in packages.items():
    try:
        importlib.import_module(module)
        print(f'✓ {pkg_name} ({module})')
    except ImportError:
        print(f'✗ {pkg_name} ({module}) — 설치 필요: pip install {pkg_name}')
```

## 3. OOP 기본 동작 확인


```python
class Greeter:
    def __init__(self, name: str):
        self.name = name

    def greet(self) -> str:
        return f'Hello, {self.name}!'


obj = Greeter('OOP')

assert obj.name == 'OOP'
assert obj.greet() == 'Hello, OOP!'
assert isinstance(obj, Greeter)

print(f'✓ 클래스 생성: {obj}')
print(f'✓ 속성 접근: obj.name = {obj.name!r}')
print(f'✓ 메서드 호출: {obj.greet()}')
print(f'✓ isinstance 확인: {isinstance(obj, Greeter)}')
```

## 4. 종합 결과


```python
import sys, importlib

checks = [
    ('Python 버전 (>= 3.9)', lambda: sys.version_info >= (3, 9)),
    ('import pytest',        lambda: importlib.import_module('pytest') or True),
    ('import bs4',           lambda: importlib.import_module('bs4') or True),
    ('import PIL (Pillow)',  lambda: importlib.import_module('PIL') or True),
    ('OOP 클래스 기본 동작',
     lambda: (lambda o: o.greet() == f'Hello, {o.name}!')(Greeter('OOP'))),
]

print('=' * 50)
print(' 환경 검증 결과')
print('=' * 50)

passed = 0
for name, check in checks:
    try:
        result = check()
        if result:
            print(f'  [PASS] {name}')
            passed += 1
        else:
            print(f'  [FAIL] {name}')
    except Exception as e:
        print(f'  [FAIL] {name} — {e}')

print('-' * 50)
print(f'결과: {passed}/{len(checks)} 통과')
if passed == len(checks):
    print('환경 구성이 완료되었습니다!')
else:
    print('위 FAIL 항목을 확인하고 설치 가이드를 참고하세요.')
print('=' * 50)
```
